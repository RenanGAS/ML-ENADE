```python
transformacaoDadosEnade = []
nome_das_novas_colunas = []
#anos_enade = [2021, 2017, 2014, 2011]
anos_enade = [2021]

definir_colunas = 1

for ano in anos_enade:
    subconjunto_ano = dados_do_enade.loc[dados_do_enade['Ano'] == ano]
    subconjunto_ano = subconjunto_ano.reset_index(drop=True)
    
    codigo_cursos_array = subconjunto_ano["Codigo_do_Curso"].unique()
    codigo_dos_cursos = codigo_cursos_array.tolist()
    
    for codigo in codigo_dos_cursos:
        subconjunto_do_codigo = subconjunto_ano.loc[subconjunto_ano['Codigo_do_Curso'] == codigo]
        
        #if subconjunto_do_codigo.iloc[0]["Curso"] != "BCC":
        #    continue
        
        subconjunto_do_codigo = subconjunto_do_codigo.reset_index(drop=True)
    
        dados_do_curso = []
    
        #if definir_colunas:
            #nome_das_novas_colunas.append('Codigo_do_Curso')
    
        #dados_do_curso.append(codigo)

        numero_notas_invalidas = 0
        numero_faltantes = 0
        numero_participantes = 0

        #for index, linha in subconjunto_do_codigo.iterrows():
        #    if subconjunto_do_codigo.at[index,"Tipo_Presenca"] == 555:
        #        numero_participantes += 1
        #    else:
        #        if subconjunto_do_codigo.at[index,"Tipo_Presenca"] == 556:
        #            numero_notas_invalidas += 1
        #        else:
        #            numero_faltantes += 1
        
        numero_notas_invalidas = sum(subconjunto_do_codigo["Tipo_Presenca"] == 556)
        numero_participantes = sum(subconjunto_do_codigo["Tipo_Presenca"] == 555)
        numero_faltantes = len(subconjunto_do_codigo) - (numero_notas_invalidas + numero_participantes)
                    
        if numero_participantes <= 1:
            continue
            
        #if definir_colunas:
            #nome_das_novas_colunas.append('Ano')
        #dados_do_curso.append(ano)
        
        numero_esperados = numero_notas_invalidas + numero_faltantes + numero_participantes
            
        if definir_colunas:
            nome_das_novas_colunas.append('Numero_Notas_Invalidas')
        dados_do_curso.append(numero_notas_invalidas / numero_esperados)
    
        if definir_colunas:
            nome_das_novas_colunas.append('Numero_Faltantes')
        dados_do_curso.append(numero_faltantes / numero_esperados)
    
        if definir_colunas:
            nome_das_novas_colunas.append('Numero_Participantes')
        dados_do_curso.append(numero_participantes / numero_esperados)
    
        for nome_coluna in subconjunto_do_codigo:
        
            if nome_coluna in atributos_do_curso:
                # if nome_coluna == 'Curso':
                #     if definir_colunas:
                #         nome_das_novas_colunas.append(nome_coluna)
                #     dados_do_curso.append(subconjunto_do_codigo[nome_coluna].iat[0])
                    
                if nome_coluna == 'Curso':
                    for valor in curso:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                        
                elif nome_coluna == 'UF':
                    for valor in uf_curso:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                        
                elif nome_coluna == 'Regiao':
                    for valor in regiao:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                        
                elif nome_coluna == 'Cat_Administrativa':
                    for valor in cat_administrativa:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                        
                elif nome_coluna == 'Cat_Org_Academica':
                    for valor in cat_org_academica:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                        
                elif nome_coluna == 'Turno':
                    #if definir_colunas:
                        #nome_das_novas_colunas.append('nulos_' + nome_coluna)
                
                    #numero_de_valores_nulos = subconjunto_do_codigo[nome_coluna].isna().sum()
                    #dados_do_curso.append(numero_de_valores_nulos)
                    
                    for valor in turno:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
    
            if nome_coluna in atributos_quantitativos:
                if definir_colunas:
                    nome_das_novas_colunas.append(nome_coluna)
            
                media_dos_valores = subconjunto_do_codigo[nome_coluna].mean()
                dados_do_curso.append(media_dos_valores)
            
            if nome_coluna in atributos_qualitativos:
                if definir_colunas:
                    nome_das_novas_colunas.append('nulos_' + nome_coluna)
                
                numero_de_valores_nulos = subconjunto_do_codigo[nome_coluna].isna().sum()
                dados_do_curso.append(numero_de_valores_nulos / numero_esperados)
        
                if nome_coluna == 'Tipo_Escola_EM':
                    for valor in tipo_escola_em:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'UF_Ensino_Medio':
                    for valor in uf_ensino_medio:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Modalidade_EM':
                    for valor in modalidade_em:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Sexo':
                    for valor in sexo:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Cor_Raca':
                    for valor in cor_raca:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Estado_Civil':
                    for valor in estado_civil:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Nl_Pai':
                    for valor in nivel_escolaridade:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Nl_Mae':
                    for valor in nivel_escolaridade:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Renda':
                    for valor in renda_familiar:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Situacao_Financeira':
                    for valor in situacao_financeira:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Moradia':
                    for valor in tipo_moradia:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Qtde_Pessoas':
                    for valor in quantidade_pessoas_moradia:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Politica_de_Ingresso':
                    for valor in politica_inclusao_social:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Situacao_Trabalho':
                    for valor in situacao_trabalho:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)

                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0

                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                    
                elif nome_coluna == 'Bolsa_de_Estudos':
                    for valor in bolsa_de_estudos:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)

                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0

                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Qtde_Livros':
                    for valor in quantidade_livros:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)

                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0

                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Horas_de_Estudo':
                    for valor in horas_estudos:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)

                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0

                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

###### Novos atributos qualitativos

                elif nome_coluna == 'Nacionalidade':
                    for valor in nacionalidade:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                
                elif nome_coluna == 'Auxilio_Permanencia':
                    for valor in auxilio_permanencia:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
                
                elif nome_coluna == 'Bolsa_Academica':
                    for valor in bolsa_academica:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Atividade_Exterior':
                    for valor in atividade_exterior:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Incentivo_Grad':
                    for valor in incentivo_grad:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Enfrentar_Diff':
                    for valor in enfrentar_diff:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Familia_Superior':
                    for valor in familia_superior:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Idioma_Estrangeiro':
                    for valor in idioma_estrangeiro:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Motivo_Curso':
                    for valor in motivo_curso:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                elif nome_coluna == 'Motivo_IES':
                    for valor in motivo_IES:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)
                        
                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()
                    
                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0
                    
                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
#######                    
                elif nome_coluna in conjunto_valores_repetidos:
                    for valor in alternativas_em_comum:
                        if definir_colunas:
                            nome_das_novas_colunas.append(valor)

                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                        if valor in valores_unicos_do_atributo:
                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                        else:
                            numero_de_ocorrencias = 0

                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

                #elif nome_coluna in segundo_conjunto_valores_repetidos:
                #    for valor in disponibilidade_dominio_professores:
                #        if definir_colunas:
                #            nome_das_novas_colunas.append(valor)

                #        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                #        if valor in valores_unicos_do_atributo:
                #            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                #        else:
                #            numero_de_ocorrencias = 0

                #        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

#                elif nome_coluna == 'Plano_de_Ensino':
#                    for valor in opiniao_plano_ensino:
#                        if definir_colunas:
#                            nome_das_novas_colunas.append(valor)

#                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

#                        if valor in valores_unicos_do_atributo:
#                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
#                        else:
#                            numero_de_ocorrencias = 0

#                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

#                elif nome_coluna == 'Formacao':
#                    for valor in avaliacao_formacao:
#                        if definir_colunas:
#                            nome_das_novas_colunas.append(valor)

#                        valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

#                        if valor in valores_unicos_do_atributo:
#                            numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
#                        else:
#                            numero_de_ocorrencias = 0

#                        dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

        transformacaoDadosEnade.append(dados_do_curso)
        definir_colunas = 0

dfDadosTransformados = pd.DataFrame(transformacaoDadosEnade)

dfDadosTransformados.columns =[nome_das_novas_colunas]
    
```
