```python
transformacaoDadosEnade = []
nome_das_novas_colunas = []

# Para definir quais edições serão incluídas
# anos_enade = [2021, 2017, 2014, 2011]
anos_enade = [2011]

definir_colunas = 1

for ano in anos_enade:
    subconjunto_ano = dados_do_enade.loc[dados_do_enade['Ano'] == ano]
    subconjunto_ano = subconjunto_ano.reset_index(drop=True)
    
    codigo_cursos_array = subconjunto_ano["Codigo_do_Curso"].unique()
    codigo_dos_cursos = codigo_cursos_array.tolist()
    
    for codigo in codigo_dos_cursos:
        subconjunto_do_codigo = subconjunto_ano.loc[subconjunto_ano['Codigo_do_Curso'] == codigo]
        
        # Para criação de conjuntos por curso de graduação
        #if subconjunto_do_codigo.iloc[0]["Curso"] != "BCC":
        #    continue
        
        subconjunto_do_codigo = subconjunto_do_codigo.reset_index(drop=True)
    
        dados_do_curso = []
        
        # Para adicionar o código do curso para webscrapping
        if definir_colunas:
            nome_das_novas_colunas.append('Codigo_do_Curso')
        dados_do_curso.append(codigo)

        numero_notas_invalidas = 0
        numero_faltantes = 0
        numero_participantes = 0
        
        numero_notas_invalidas = sum(subconjunto_do_codigo["Tipo_Presenca"] == 556)
        numero_participantes = sum(subconjunto_do_codigo["Tipo_Presenca"] == 555)
        numero_faltantes = len(subconjunto_do_codigo) - (numero_notas_invalidas + numero_participantes)
                    
        if numero_participantes <= 1:
            continue
        
        # Para adicionar o ano da edição
        #if definir_colunas:
            #nome_das_novas_colunas.append('Ano')
        #dados_do_curso.append(ano)
        
        numero_esperados = numero_notas_invalidas + numero_faltantes + numero_participantes
            
        if definir_colunas:
            nome_das_novas_colunas.append('Numero_Notas_Invalidas')
        dados_do_curso.append(numero_notas_invalidas / numero_esperados)
    
        if definir_colunas:
            nome_das_novas_colunas.append('Numero_Faltantes')
        dados_do_curso.append(numero_faltantes / numero_esperados)
    
        if definir_colunas:
            nome_das_novas_colunas.append('Numero_Participantes')
        dados_do_curso.append(numero_participantes / numero_esperados)
    
        for nome_coluna in subconjunto_do_codigo:

            if nome_coluna in atributos_curso:

                # Para adicionar o nome do curso
                # if nome_coluna == 'Curso':
                #     if definir_colunas:
                #         nome_das_novas_colunas.append(nome_coluna)
                #     dados_do_curso.append(subconjunto_do_codigo[nome_coluna].iat[0])

                for valor in dict_quest_altern[nome_coluna]:
                    if definir_colunas:
                        nome_das_novas_colunas.append(valor)

                    valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                    if valor in valores_unicos_do_atributo:
                        numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                    else:
                        numero_de_ocorrencias = 0

                    dados_do_curso.append(numero_de_ocorrencias / numero_esperados)
    
            if nome_coluna in atributos_continuos:
                if definir_colunas:
                    nome_das_novas_colunas.append(nome_coluna)
            
                media_dos_valores = subconjunto_do_codigo[nome_coluna].mean()
                dados_do_curso.append(media_dos_valores)
            
            if nome_coluna in atributos_categoricos:
                if definir_colunas:
                    nome_das_novas_colunas.append('nulos_' + nome_coluna)
                
                numero_de_valores_nulos = subconjunto_do_codigo[nome_coluna].isna().sum()
                dados_do_curso.append(numero_de_valores_nulos / numero_esperados)
                    
                for valor in dict_quest_altern[nome_coluna]:
                    if definir_colunas:
                        nome_das_novas_colunas.append(valor + "_" + nome_coluna)

                    valores_unicos_do_atributo = subconjunto_do_codigo[nome_coluna].dropna().unique()

                    if valor in valores_unicos_do_atributo:
                        numero_de_ocorrencias = subconjunto_do_codigo[nome_coluna].value_counts()[valor]
                    else:
                        numero_de_ocorrencias = 0

                    dados_do_curso.append(numero_de_ocorrencias / numero_esperados)

        transformacaoDadosEnade.append(dados_do_curso)
        definir_colunas = 0

dfDadosTransformados = pd.DataFrame(transformacaoDadosEnade)

dfDadosTransformados.columns =[nome_das_novas_colunas]
    
```
