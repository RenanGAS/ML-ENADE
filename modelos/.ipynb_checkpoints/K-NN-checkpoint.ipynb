```python
from sklearn.neighbors import (NeighborhoodComponentsAnalysis, KNeighborsClassifier)
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

import pandas as pd
import numpy as np

from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV

import graphviz 
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import top_k_accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("../tabela_final_treinamento.csv")

X = data.iloc[:, 0:225]
Y = data.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, stratify=Y, test_size=0.20, random_state=42)
```

```python
scaler_train = StandardScaler()
```

```python
scaler_train.fit(X_train)
normalized_train = scaler_train.transform(X_train)
normalized_train
```

```python
scaler_test = StandardScaler()
```

```python
scaler_test.fit(X_test)
normalized_test = scaler_test.transform(X_test)
normalized_test
```

```python
param_grid = {"n_neighbors": [3, 5, 10], 
              "weights": ["uniform", "distance"],
              "metric": ["euclidean", "manhattan", "nan_euclidean"]
             }
```

```python
classfier = KNeighborsClassifier()
```

```python
search = HalvingGridSearchCV(classfier, param_grid, scoring="f1_weighted").fit(X_train, y_train) 
```

```python
search_normalized = HalvingGridSearchCV(classfier, param_grid, scoring="f1_weighted").fit(normalized_train, y_train) 
```

```python
search.score(X_test, y_test)
```

```python
search_normalized.score(normalized_test, y_test)
```

```python
y_score = search.predict_proba(X_test)
```

```python
top_k_accuracy_score(y_test, y_score, k=2, labels=[1, 2, 3, 4, 5])
```

```python
y_pred_HGS = search.predict(X_test)
```

```python
cm_HGS = confusion_matrix(y_test, y_pred_HGS, labels=search.classes_)
disp_HGS = ConfusionMatrixDisplay(confusion_matrix=cm_HGS,
                              display_labels=search.classes_)
disp_HGS.plot()
plt.show()
```

```python
search.best_params_
```

```python
nca = NeighborhoodComponentsAnalysis(random_state=42)

classfier = KNeighborsClassifier()
knnClassfier = HalvingGridSearchCV(classfier, param_grid, scoring="f1_weighted")

# knnClassfier = KNeighborsClassifier(metric="euclidean", n_neighbors=10, weights="distance")

nca_pipe = Pipeline([('nca', nca), ('knnClassfier', knnClassfier)])
nca_pipe.fit(X_train, y_train)
```

```python
nca_pipe.score(X_test, y_test)
```

```python
y_pred_NCA = nca_pipe.predict(X_test)
```

```python
cm_NCA = confusion_matrix(y_test, y_pred_NCA, labels=nca_pipe.classes_)
disp_NCA = ConfusionMatrixDisplay(confusion_matrix=cm_NCA,
                              display_labels=nca_pipe.classes_)
disp_NCA.plot()
plt.show()
```
